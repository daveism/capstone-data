```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

datadf = pd.read_csv('data/data-and-metrics.csv')

def group(column, df):
    grouped = df.groupby([column]).agg({'timetaken': ['mean'],'correct': ['mean']})
    grouped = grouped.reset_index()
    grouped.columns = [column,'timetaken_mean', 'correct_mean']
    return grouped
```

```python
chart = sns.FacetGrid(datadf, col="target", hue="target",
                  col_order=["gestalt","less gestalt","no target"],
                  height=5, aspect=1)
chart.map(sns.histplot, "timetaken", bins=20 )
```

```python
chart = sns.FacetGrid(datadf, col="target", row='basemap', hue="target",
                  col_order=["gestalt", "less gestalt", "no target"],
                  row_order=["none", "dark", "imagery","streets"],
                  height=5, aspect=1)
chart.map(sns.histplot, "timetaken", bins=20 )
```

```python
chart = sns.FacetGrid(datadf, col="target", row='color', hue="target",
                  col_order=["gestalt", "less gestalt", "no target"],
                  row_order=["blue","mixed","red","matches"],
                  height=5, aspect=1)
chart.map(sns.histplot, "timetaken", bins=20 )
```

```python
dflim = datadf.loc[datadf['distractor'] == 'random']


chart = sns.FacetGrid(dflim, col="target", row='distractorlevel', hue="target",
                  col_order=["gestalt", "less gestalt", "no target"],
                  row_order=["low","medium","high","all"],
                  height=5, aspect=1)
chart.map(sns.histplot, "timetaken", bins=20 )
```
