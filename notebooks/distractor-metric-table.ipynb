```python
import pandas as pd
import numpy as np
datadf = pd.read_csv('data/metrics.csv')

```

```python

datadflim = datadf.loc[datadf['distractormet'] == 'random']
datadflim = datadflim.loc[datadf['basemapmet'] == 'none']
grp = datadflim.groupby(['colormet', 'distractorlevelmet']).agg({
     'distractorlevelmet': ['count'],
    'number_of_patches': ['min'],
    'area_mn': ['min'],
    'shape_index_mn': ['min'],
    'euclidean_nearest_neighbor_mn': ['min'],
    'contagion': ['min']
})
grp = grp.reset_index()
grp.columns = ['color', 'distractorlevel', 'count', 'nop_mean', 'aap_mean', 'lsi_mean',  'enn_mean', 'cont_mean']
grp = grp.pivot('color', 'distractorlevel', ['count', 'nop_mean', 'aap_mean', 'lsi_mean', 'enn_mean', 'cont_mean'])
grp.to_csv('data/distractor-random-table.csv')
grp
```

```python
datadf = pd.read_csv('data/metrics.csv')
datadflim = datadf.loc[datadf['distractormet'] == 'clustered']
datadflim = datadflim.loc[datadf['basemapmet'] == 'none']
grp = datadflim.groupby(['colormet', 'distractorlevelmet']).agg({
    'distractorlevelmet': ['count'],
    'number_of_patches': ['min'],
    'area_mn': ['min'],
    'shape_index_mn': ['min'],
    'euclidean_nearest_neighbor_mn': ['min'],
    'contagion': ['min']
})
grp = grp.reset_index()
grp.columns = ['color', 'distractorlevel', 'count', 'nop_mean', 'aap_mean', 'lsi_mean',  'enn_mean', 'cont_mean']
grp = grp.pivot('color', 'distractorlevel', ['count', 'nop_mean', 'aap_mean', 'lsi_mean', 'enn_mean', 'cont_mean'])
column_order = ["large smooth","large rough","medium smooth","medium rough","small smooth","small rough"]
grp = grp.reindex(column_order, level=1, axis=1)
grp.to_csv('data/distractor-clustered-table.csv')
grp
```
